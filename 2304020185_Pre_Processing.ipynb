## CEK DAN PENANGANAN OUTLIER

**1. Boxplot**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Kolom numerik pada dataset Covid-19 yang ingin di cek outliernya
num_cols = ['new_tested', 'acc_tested', 'new_confirmed', 'acc_confirmed',
            'acc_negative', 'being_checked', 'isolated',
            'new_released', 'acc_released', 'new_deceased', 'acc_deceased']

# Visualisasi Boxplot
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(y=data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

Berdasarkan output di atas, visualisasi boxplot digunakan untuk menunjukkan sebaran data serta mendeteksi keberadaan nilai ekstrem (outlier) pada setiap variabel numerik dalam dataset Covid-19 Indonesia, yaitu new_tested, acc_tested, new_confirmed, acc_confirmed, acc_negative, being_checked, isolated, new_released, acc_released, new_deceased, dan acc_deceased. Terlihat bahwa beberapa variabel seperti new_tested, acc_tested, new_confirmed, acc_confirmed, acc_negative, isolated, dan acc_deceased memiliki nilai outlier yang ditunjukkan dengan titik di luar garis whisker, yang berarti terdapat nilai yang jauh lebih tinggi dibandingkan sebagian besar data lainnya. Hal ini mengindikasikan adanya lonjakan jumlah tes, kasus terkonfirmasi, maupun kematian pada waktu tertentu. Sementara itu, variabel seperti being_checked, new_released, acc_released, dan new_deceased cenderung memiliki sebaran data yang lebih sempit dan relatif stabil dengan jumlah outlier yang lebih sedikit. Secara keseluruhan, hasil visualisasi ini menunjukkan bahwa data memiliki variasi yang cukup tinggi antar variabel

**2. IQR**

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # Kuartil 1
    Q3 = data[column].quantile(0.75)  # Kuartil 3
    IQR = Q3 - Q1  # Rentang interkuartil
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Cek jumlah outlier untuk setiap kolom numerik
for col in num_cols:
    outliers = detect_outliers_iqr(data, col)
    print(f"{col}: {len(outliers)} outliers")
```

Berdasarkan output yang di atas, dilakukan proses deteksi outlier pada setiap variabel numerik dalam dataset Covid-19 Indonesia, yaitu new_tested, acc_tested, new_confirmed, acc_confirmed, acc_negative, being_checked, isolated, new_released, acc_released, new_deceased, dan acc_deceased, menggunakan metode Interquartile Range (IQR). Metode ini mengidentifikasi nilai yang berada di luar batas bawah (Q1 − 1.5×IQR) dan batas atas (Q3 + 1.5×IQR) sebagai outlier. Hasil yang diperoleh menunjukkan jumlah outlier pada masing-masing kolom, sehingga dapat diketahui variabel mana saja yang memiliki data ekstrem dan perlu diperhatikan dalam proses analisis lebih lanjut.

**3. Winsorizing (Mengganti Outlier dengan Batas IQR)**

```python
def winsorize_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])

# Terapkan Winsorizing
for col in num_cols:
    winsorize_iqr(data, col)
```

Blok diatas melakukan proses penanganan outlier pada seluruh variabel numerik dalam dataset Covid-19 Indonesia, yaitu new_tested, acc_tested, new_confirmed, acc_confirmed, acc_negative, being_checked, isolated, new_released, acc_released, new_deceased, dan acc_deceased, menggunakan metode winsorizing berbasis Interquartile Range (IQR). Metode ini bekerja dengan membatasi nilai yang berada di luar batas bawah (Q1 − 1.5×IQR) dan batas atas (Q3 + 1.5×IQR), dengan cara menggantinya menjadi nilai batas tersebut. Dengan demikian, nilai ekstrem tidak dihapus, tetapi disesuaikan agar tidak terlalu memengaruhi hasil analisis selanjutnya.
